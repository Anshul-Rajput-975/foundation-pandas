```python
# 4. Indexing and Selection
import pandas as pd
data = pd.DataFrame({
'Park': ['A', 'A', 'B', 'B'],
'Name': ['Anna','Ben','Chris','Daisy'],
'Lemonades': [2, 3, 4, 5]
})
data.set_index("Name",inplace=True) #set_index() works with columns
data
```

```python
multi_data = pd.DataFrame({
'Park': ['A', 'A', 'B', 'B'],
'Name': ['Anna','Ben','Chris','Daisy'],
'Lemonades': [2, 3, 4, 5]
}).set_index(['Park', 'Name'])
multi_data #multi-index
```

```python
# Selecting with Conditions
data[data["Lemonades"]>2]
```

```python
# Query Method
data.query("Park=='A'")
```
