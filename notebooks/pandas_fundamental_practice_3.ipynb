```python
# String Operations
import pandas as pd
data = pd.DataFrame({
'Park': ['A', 'A', 'B', 'B'],
'Name': ['Anna','Ben','chris','daisy'],
'Lemonades': [2, 3, 4, 5]
})
data["Name"]=data["Name"].str.capitalize()
data
```

```python
#  Data Aggregation and Grouping
grouped=data.groupby("Name") # You are grouping the rows of the DataFrame data based on unique values in the Name column.
grouped # groupby object
for name,group in grouped:
    print("Group:", name)
    print(group)
    print()
```

```python
grouped.sum()
```

```python
grouped["Lemonades"].mean() # select only the Lemonades column inside each group and compute mean
```

```python
# Pivot Tables
import pandas as pd
data = pd.DataFrame({'Park': ['A', 'A', 'B', 'B'],
'Name': ['Anna','Ben','chris','daisy'],
'Lemonades': [2, 3, 4, 5]
})
# pivot_table() is used to summarize data by grouping rows and applying an aggregation (mean, count, sum, etc.).
"""values → column you want to analyze
index → row-wise grouping (like groupby)
columns → column-wise grouping 
aggfunc → how to summarize (mean, sum, count, etc.)"""
data.pivot_table(values=["Park","Lemonades"],index="Name", aggfunc="first")
```

```python
# Crosstab 
pd.crosstab(data["Name"],data["Lemonades"])
```
