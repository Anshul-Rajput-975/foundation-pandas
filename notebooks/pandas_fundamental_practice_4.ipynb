```python
# Merging, Joining and Concatenating
import pandas as pd
data = pd.DataFrame({'Park': ['A', 'A', 'B', 'B'],
'Name': ['Anna','Ben','chris','daisy'],
'Lemonades': [2, 3, 4, 5]
})
data2 = pd.DataFrame({'Store': ['X', 'Z', 'Y', 'Y'],
    'Employee': ['Ravi', 'Aman', 'Neha', 'Pooja'],
    'Sales': [120, 150, 200, 180]
})
pd.concat([data,data2],ignore_index=True)
```

```python
data.append(data2) # removed in pandas 2.0
```

```python
# common column is required to merge them
data2 = pd.DataFrame({'Store': ['X', 'Z', 'Y', 'Y'],
    'Name': ['Anna','Ben','chris','daisy'],
    'Sales': [120, 150, 200, 180]
})
pd.merge(data,data2,on="Name") 
```

```python
# Join on Index
data.join(data2.set_index("Name"),on="Name") # The right DataFrame must have that column as its index
```

```python
#  Working with Time Series
data3 = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'Sales': [100, 150, 130, 170],
    'City': ['Indore', 'Bhopal', 'Indore', 'Bhopal']
})
data3
```

```python
data3["Date"]=pd.to_datetime(data3["Date"])
data3
```

```python
data4 = pd.DataFrame({
    'Date': [
        '2024-01-01', '2024-01-02', '2024-01-03',
        '2024-01-04', '2024-01-05', '2024-01-06',
        '2024-01-07'
    ],
    'Lemonades': [10, 15, 20, 18, 25, 30, 28]
})
data4["Date"]=pd.to_datetime(data4["Date"])
data4.set_index("Date",inplace=True)
```

```python
data4.resample('W').sum()
```

```python
data4['Previous Lemonades'] = data4['Lemonades'].shift(4)
data4
```

```python
data4['Rolling Average'] = data4['Lemonades'].rolling(window=3).mean()
data4
```
