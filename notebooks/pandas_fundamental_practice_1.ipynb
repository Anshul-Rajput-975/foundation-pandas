```python
import pandas as pd
s1 = pd.Series(data=[1,2,3,4,5],index=["a","b","c","d","e"])
s2 = pd.Series(data=[10,20,30,40,50],index=["a","b","c","d","f"])
s3 = pd.Series(data=["a","b","a","d","f"],index=["a","b","c","d","f"])
data=pd.DataFrame({"c1":s1,"c2":s2,"c3":s3})
data.loc[data["c3"]=="a","c2"]=5
data
```

```python
data.rename(columns={"c1":"column"})
data
```

```python
del data["c3"]
data
```
