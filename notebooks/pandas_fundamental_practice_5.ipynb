```python
# Data Visualization
# Basic Plotting
import pandas as pd

data = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=7, freq='D'),
    'Lemonades': [10, 15, 20, 18, 25, 30, 28],
    'Temperature': [20, 22, 25, 24, 26, 28, 27]
})
data['Lemonades'].plot()
```

```python
data["Lemonades"].plot(kind="bar")
```

```python
data["Lemonades"].plot(kind="line")
```

```python
data["Lemonades"].plot(kind="barh")
```

```python
data["Lemonades"].plot(kind="hist",bins=2)
```

```python
data["Lemonades"].plot(kind="box")
```

```python
import pandas as pd
data = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=7, freq='D'),
    'Lemonades': [10, 15, 20, 18, 25, 30, 28],
    'Temperature': [20, 22, 25, 24, 26, 28, 27]
})
data.plot(kind='scatter', x='Temperature', y='Lemonades')
```

```python
data['Lemonades'].plot(kind='area')
```

```python
data['Lemonades'].plot(kind='pie')
```

```python
data['Lemonades'].plot(
    title='Lemonade Sales Over Time',
    xlabel='Date',
    ylabel='Lemonades Sold'
)
```

```python
"""scatter  must use DataFrame, not Series
pie  usually used for category proportion
Index is automatically used as X-axis
Pandas plots internally use matplotlib"""
```
